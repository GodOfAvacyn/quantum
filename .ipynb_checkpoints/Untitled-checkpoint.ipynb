```python
from random import randint
import numpy as np
```

```python
def gcd(a,b):
    """a can either be a huge power number broken up like x ^ y + z in the form (x,y,z), or a number"""
    while b != 0:
        t = b
        if type(a) == tuple:
            b = (pow(a[0],a[1],b)+a[2]) % b
        else:
            b = a%b
        a = t
    return a
```

```python
def classical_subroutine(a,N):
    if gcd(a,N) != 1:
        return -1
    r = 1
    t = a
    while t%N != 1:
        t = (a*t)%N
        r += 1
    return r
print(classical_subroutine(13,1532))
```

```python
def factor(N):
    while True:
        a = randint(2,N-1)
        K = gcd(a,N)
        if K != 1:
            return (K,N/K)
        r = classical_subroutine(a,N)
        if r%2 == 1:
            continue
        g = gcd((a,int(r/2),1),N)
        if g != 1 and g != N:
            return (g, int(N/g))
```
