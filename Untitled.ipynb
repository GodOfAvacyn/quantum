```python
def gcd(a,b):
    """a can either be a huge power number broken up like x ^ y + z in the form (x,y,z), or a number"""
    while b != 0:
        t = b
        if type(a) == tuple:
            b = (pow(a[0],a[1],b)+a[2]) % b
        else:
            b = a%b
        a = t
    return a
```

```python
def classical_subroutine(a,N):
    if gcd(a,N) != 1:
        return -1
    r = 1
    t = a
    while t%N != 1:
        t = (a*t)%N
        r += 1
    return r
print(classical_subroutine(13,1532))
```

```python
def factor(N):
    while True:
        a = randint(2,N-1)
        K = gcd(a,N)
        if K != 1:
            return (K,N/K)
        r = classical_subroutine(a,N)
        if r%2 == 1:
            continue
        g = gcd((a,int(r/2),1),N)
        if g != 1 and g != N:
            return (g, int(N/g))
```

## Quantum Fourier Transform
The classical Fourier transform is an operator which takes a periodic function as input and outputs another
function, with maxima corresponding to the period of the input.
$$ \text{Fourier}[f(x)]  = \int_{t_1}^{t_2} f(t)\cdot e^{i\theta x t } dt$$
The more time between $t_1$ and $t_2$, the more accurate the results will be.

The quantum Fourier transform is similar. For a quantum state with $n$ qubits:
$$ \text{QFT}(|x\rangle) = \frac{1}{\sqrt{N}}\sum_{y=0}^{N-1} e^{\frac{2\pi ix \cdot y}{N}}|x\rangle \quad (N=2^n)$$
The $\frac{1}{\sqrt{N}}$ term is for normalization, since we want a unitary operator. Anyways, we can simplify this further when we consider that each $y$ is a quantum state itself, where $y=5$ means $y$ is the state of the binary representation of 5 ($|101\rangle$).
So $y = 2^{n-1}y_1 + 2^{n-2}y_2 \dots + y_n = \sum_{k=1}^n 2^{n-k}y_k  = \sum_{k=1}^n \frac{2^ny_k}{2^k} = \sum_{k=1}^n \frac{Ny_k}{2^k}$. Substituting this in for $y$ above, and using the $a^(b+c) = a^ba^c$ rule of exponentials, we get

$$ \frac{1}{\sqrt{N}}\sum_{y=0}^{N-1} \prod_{k=0}^n e^{\frac{2\pi ix \cdot Ny_k}{2^kN}}|x\rangle \quad = $$
$$ \frac{1}{\sqrt{N}}\sum_{y=0}^{N-1} \prod_{k=0}^n e^{\frac{2\pi ix \cdot y_k}{2^k}}|x\rangle \quad$$
We can break these products up into:
$$ \frac{1}{\sqrt{N}} (|0\rangle + e^{\frac{2\pi ix}{2^1}}|1\rangle) \otimes (|0\rangle + e^{\frac{2\pi ix}{2^2}}|1\rangle) \dots \otimes (|0\rangle + e^{\frac{2\pi ix}{2^n}}|1\rangle)$$
This tensor product is just a series of controlled rotations, each with a phase of $pi / 2^{k-1}$ for some k!

```python
def QFT(n):
    circuit = QuantumCircuit(n)
    for qubit in range(n):
        circuit.h(qubit)
        for control in range(qubit+1,n):
            circuit.cp(np.pi / (2**(control - qubit)),control,qubit)
    circuit.name = "QFT"
    return circuit
display(QFT(5).draw(output="mpl"))
```

## Phase Estimation
A quantum unitary operator's eigenvalues must always be of the form $e^{i\theta}$, since numbers of this form are the only valid scaling factors that preserve unit length. One important quantum question is, for a unitary operator $U$ with eigenvalue $\lambda_u = e^{i\theta}$, is it possible to extract theta out?

A solution may be to measure a quantum state ~100 times and then use the probabilities as a measure of what phase is being applied to a state. But for the $|+ \rangle$ state, the probability of measuring the 0 & 1 states are both 0.5, and for the $|- \rangle$ state, the probabilities are the same. This means that measuring probability while running a quantum state over & over is not enough.

We can actually use QFT here to help out: what we do to find the phase of some unitary operator $U$ is we prepare two quantum registers, one acting as a control qubit & the others as the input for the operator $U$. We force the control into the equilibrium state with a Hadamard gate, then apply a controlled version of $U$ onto it, then we apply the hadamard once more. For a unitary operator with phase $e^{i\theta}$, and an input state $|\psi \rangle$, our control qubit now looks like this:
$$ \frac{1}{\sqrt2} \bigg(\frac{ |0\rangle + |1\rangle }{\sqrt{2}}\bigg) |\psi + e^{i\theta}\bigg(\frac{ |0\rangle - |1\rangle }{\sqrt{2}}\bigg) |\psi\rangle$$
$$ = \frac{1}{\sqrt2} \bigg[ |0\rangle \bigg( 1 + e^{i\theta} \bigg) \bigg] + |1\rangle \bigg( 1 - e^{i\theta} \bigg) \bigg] |\psi\rangle$$

The probability of measuring the 0 state is $|| \frac{1}{\sqrt2}(1 + e^{i\theta})||$, and almost vice versa for the 1 state (with a "-").

This should be enough math to figure out exactly what the paramter $\theta$ is. We can increase the accuracy of our findings by increasing how many bits we use for a control, & for each of them using more & more repeated applications of the operator $U$. Our output of such a circuit would be Fourier Transform itself, so applying an inverse QFT on our control qubits will get us  usable information.

Here is an example where $U$ is a simple phase of 0.5.

```python
def Phase(n):
    circuit = QuantumCircuit(2*n)
    for i in range(n):
        circuit.h(i)
    for k in range(n):
        for qubit in range(n,2*n):
            theta = 0.5*(2**(n-k-1))  # each time we compound the number of times we apply the gate.
            circuit.cp(theta,k,qubit)
    return circuit.compose(QFT(n).inverse(), qubits=[j for j in range(n)])
    


display(Phase(3).draw(output="mpl"))
```

## Modular exponentiation 
To perform phase estimation on a modular exponentiation function, we a unitary operator which a function
$f_{(a,N)}(x) = a^x\mod N$. Here is that circuit for $a = 4$, $N = 21$.

The clever authors of this paper actually aready know the stupid period here: 3. $4^3 = 64, 64 = 3*21 + 1$. Therefore the construction of this version of shore's algorithm is completely meaningless & a waste of time.

The only possible mappings for $U$ are $|1\rangle \rightarrow |4\rangle, |4\rangle \rightarrow |16\rangle, |16\rangle \rightarrow |1\rangle$.
For $U^2$ (the gate applied twice, as needed by QPE) is
$|1\rangle \rightarrow |16\rangle, |4\rangle \rightarrow |1\rangle, |16\rangle \rightarrow |4\rangle$
For $U^4$, the final one we need is: $|1\rangle \rightarrow |4\rangle, |4\rangle \rightarrow |16\rangle, |16\rangle \rightarrow |1\rangle$.

Rewritten in binary state form, and only working with the $\log_4$ of numbers 1, 4, 16, aka 0, 1, 2, and disregarding our domains are:
$$ U^{2^0} = 00 \rightarrow 01 \quad 01 \rightarrow 10 \quad 10 \rightarrow 00$$
$$ U^{2^1} = 00 \rightarrow 10 \quad 01 \rightarrow 00 \quad 10 \rightarrow 01$$
$$ U^{2^2} = 00 \rightarrow 01 \quad 01 \rightarrow 10 \quad 10 \rightarrow 00$$

We also need to adjust each of these gates by controlling them with an additional qubit (it's not just a $U$ gate, it's a controlled-U).
So all we need now are 3 circuits, each that take a control qubit & 2 working qubits, & we'll have all our circuits necessary for this utterly useless implementation of Shor's algorithm. Notice that U0 & U2 are the same. Also notice that U1 clearly is just U2 twice.

```python
circ = QuantumCircuit(3)
def U(n):
    circ = QuantumCircuit(3)
    for i in range(n):
        circ.x(2)
        circ.toffoli(0,2,1)
        circ.x(2)
        circ.cnot(2,1)
        circ.toffoli(0,1,2)
        circ.cnot(2,1)
    return circ
Q0 = QuantumCircuit(3)
Q0.cnot(0,2)
U(1).draw(output="mpl")  # verification
```
